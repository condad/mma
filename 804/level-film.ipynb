```python
# Plot 1/8th Kelly vs risk-free rate and 1/4th Kelly in a panel below
import numpy as np
import matplotlib.pyplot as plt

# Domain: x (risk-free rate) range starts at 0. Set upper bound to 3%.
r_min = 0.0
r_max = 0.03
x = np.linspace(r_min, r_max, 500)

# Formulas
y1 = ((0.0217 - x) / (0.0953**2)) / 8.0   # 1/8th Kelly
y2 = ((0.0217 - x) / (0.0953**2)) / 4.0   # 1/4th Kelly

# Zero-crossing
x_zero = 0.0217
y_zero1 = ((0.0217 - x_zero) / (0.0953**2)) / 8.0
y_zero2 = ((0.0217 - x_zero) / (0.0953**2)) / 4.0

fig, axes = plt.subplots(2, 1, sharex=True, figsize=(8, 8))

# Top panel: 1/8th Kelly
ax = axes[0]
ax.plot(x, y1, lw=2, label='1/8th Kelly')
ax.scatter([x_zero], [y_zero1], color='red', zorder=5)
ax.annotate(f'zero crossing\nx={x_zero:.4f}', xy=(x_zero, y_zero1), xytext=(x_zero + 0.001, y_zero1 + 0.08),
            arrowprops=dict(arrowstyle='->', color='red'), color='red')
ax.axhline(0, color='gray', lw=0.7)
ax.set_ylabel('1/8th kelly criterion')
ax.set_title('1/8th Kelly Criterion vs Risk-free Rate')
ax.grid(True)
ax.legend()

# Bottom panel: 1/4th Kelly
ax2 = axes[1]
ax2.plot(x, y2, lw=2, color='tab:orange', label='1/4th Kelly')
ax2.scatter([x_zero], [y_zero2], color='red', zorder=5)
ax2.annotate(f'zero crossing\nx={x_zero:.4f}', xy=(x_zero, y_zero2), xytext=(x_zero + 0.001, y_zero2 + 0.15),
             arrowprops=dict(arrowstyle='->', color='red'), color='red')
ax2.axhline(0, color='gray', lw=0.7)
ax2.set_xlabel('Risk-free Interest Rates')
ax2.set_ylabel('1/4th kelly criterion')
ax2.set_title('1/4th Kelly Criterion vs Risk-free Rate')
ax2.grid(True)
ax2.legend()

plt.tight_layout()
plt.show()
```
