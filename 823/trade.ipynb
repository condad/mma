```python
# Load Required Libraries
import pandas as pd

# Load tweet_metadata and tweets CSV Files
tweet_metadata = pd.read_csv('tweet_metadata.csv')
tweets = pd.read_csv('tweets.csv')

# Join DataFrames on "id" Column
# If either file uses a different column name for the tweet ID, adjust as needed
df = pd.merge(tweet_metadata, tweets, on='id', how='inner')

# Display the first few rows of the joined DataFrame
df.head()
```

```python
# Load FX Calls and FX Puts datasets
fx_calls = pd.read_csv('fx_calls.csv')
fx_puts = pd.read_csv('fx_puts.csv')

# Display the first few rows of each DataFrame
print('FX Calls:')
display(fx_calls.head())
print('FX Puts:')
display(fx_puts.head())
```

```python
# Filter tweets: adjusted sentiment thresholds and specific currencies (CAD, INR, JPY, GBP)
# Fix: properly handle currencies_found stored as string representations of lists (e.g., "['JPY','USD']")
# Criteria:
#  - sentiment_compound > 0.5 (strong positive)
#    OR sentiment_compound < -0.5 (strong negative)
#  - currencies_found (parsed) contains any of target currencies

target_currencies = {'CAD','INR','JPY','GBP'}

import ast

# Ensure required columns exist
missing_cols = [col for col in ['sentiment_compound','currencies_found'] if col not in df.columns]
if missing_cols:
    print(f"Warning: Missing columns in df: {missing_cols}. Attempting to proceed with available data.")

# Normalize currencies_found to list
def normalize_currencies(val):
    if isinstance(val, list):
        return val
    if pd.isna(val):
        return []
    if isinstance(val, str):
        s = val.strip()
        # Attempt to parse list-like string
        if s.startswith('[') and s.endswith(']'):
            try:
                parsed = ast.literal_eval(s)
                if isinstance(parsed, list):
                    return [str(x).strip().upper() for x in parsed]
            except Exception:
                pass
        # Fallback: split by commas
        return [piece.strip().upper() for piece in s.split(',') if piece.strip()]
    return []

if 'currencies_found' in df.columns:
    norm_col = df['currencies_found'].apply(normalize_currencies)
else:
    norm_col = pd.Series([[]]*len(df))

# Helper to check currency presence
def has_target_currency_list(lst):
    return any(cur in target_currencies for cur in lst)

if 'sentiment_compound' in df.columns:
    sentiment_mask = (df['sentiment_compound'] > 0.5) | (df['sentiment_compound'] < -0.5)
else:
    sentiment_mask = pd.Series([False]*len(df))

currency_mask = norm_col.apply(has_target_currency_list)

filtered_df = df[sentiment_mask & currency_mask].copy()
filtered_df['normalized_currencies'] = norm_col[sentiment_mask & currency_mask]

print(f"Total tweets: {len(df)}")
print(f"Tweets with sentiment >0.5 or < -0.5 mentioning target currencies: {len(filtered_df)}")

# Debug: strong positive JPY candidates
if 'sentiment_compound' in df.columns:
    debug_candidates = df[(df['sentiment_compound'] > 0.5) & norm_col.apply(lambda L: 'JPY' in L)]
    if len(debug_candidates) > 0:
        print(f"Strong positive JPY candidates total: {len(debug_candidates)}")

# Show sample rows
if len(filtered_df) > 0:
    show_cols = []
    for c in ['id','tweet_id','date','text','sentiment_compound','currencies_found','normalized_currencies']:
        if c in filtered_df.columns:
            show_cols.append(c)
    display(filtered_df.head(20)[show_cols])
else:
    print("No records matched the >0.5/<-0.5 sentiment criteria with target currencies.")
```

```python
# Create mover_tweets dataframe from filtered results
mover_tweets = filtered_df.copy()
print(f"mover_tweets created with {len(mover_tweets)} rows.")

# Basic per-currency breakdown
if len(mover_tweets) > 0 and 'normalized_currencies' in mover_tweets.columns:
    from collections import Counter
    flat = [cur for lst in mover_tweets['normalized_currencies'] for cur in lst if cur in target_currencies]
    freq = Counter(flat)
    print("Per target currency counts:")
    for cur, cnt in freq.items():
        print(f"  {cur}: {cnt}")

# Show a concise sample
sample_cols = [c for c in ['id','tweet_id','date','text','sentiment_compound','normalized_currencies'] if c in mover_tweets.columns]
if sample_cols:
    display(mover_tweets.head(10)[sample_cols])
```

```python
# Long Strangle Backtest (simplified): fixed premium 0.002 per leg, forward-filled FX rates, widened lookback.
# Removed dynamic premium/target return logic; this cell now uses a constant premium.

import os, math, datetime as dt
import pandas as pd

FIXED_PREMIUM_PER_LEG = 0.002
FILL_MISSING_RATES = True
RATE_LOOKBACK_DAYS = 10

# Column candidate mappings
col_candidates = {
    'date': ['TradeDate','Date','trade_date','listing_date'],
    'expiry': ['Expiration','Expiry','expiration','expiry_date','maturity'],
    'underlying': ['Underlying','underlying','Currency','symbol'],
    'strike': ['Strike','strike','StrikePrice'],
    'bid': ['Bid','bid'],
    'ask': ['Ask','ask'],
    'mid': ['Mid','mid','Price','premium']
}

def map_columns(df, label):
    mapping = {}
    lower = {c.lower(): c for c in df.columns}
    for key, cands in col_candidates.items():
        for cand in cands:
            if cand.lower() in lower:
                mapping[key] = lower[cand.lower()]
                break
    missing = [k for k in ['date','expiry','underlying','strike'] if k not in mapping]
    if missing:
        print(f"Warning: {label} missing columns needed for selection: {missing}")
    return mapping

calls_map = map_columns(fx_calls, 'fx_calls')
puts_map  = map_columns(fx_puts,  'fx_puts')

# Parse dates
for mp, df_opts in [(calls_map, fx_calls),(puts_map, fx_puts)]:
    if 'date' in mp:
        df_opts[mp['date']] = pd.to_datetime(df_opts[mp['date']], errors='coerce', utc=True).dt.tz_localize(None)
    if 'expiry' in mp:
        df_opts[mp['expiry']] = pd.to_datetime(df_opts[mp['expiry']], errors='coerce', utc=True).dt.tz_localize(None)

exchange_dir = 'exchange_rates'
rate_cache = {}
rate_col_cache = {}

def load_rates(cur):
    if cur in rate_cache:
        return rate_cache[cur]
    path = os.path.join(exchange_dir, f"{cur}_USD_rates.csv")
    if not os.path.exists(path):
        print(f"Rate file missing for {cur}: {path}")
        rate_cache[cur] = None
        return None
    df_r = pd.read_csv(path)
    date_col = None
    for c in df_r.columns:
        if c.lower() in ('date','day') or 'date' in c.lower():
            date_col = c; break
    if date_col is None:
        date_col = df_r.columns[0]
    df_r[date_col] = pd.to_datetime(df_r[date_col], errors='coerce', utc=True).dt.tz_localize(None)
    rate_col = None
    for c in df_r.columns:
        if c == date_col: continue
        if pd.api.types.is_numeric_dtype(df_r[c]):
            rate_col = c; break
    if rate_col is None:
        print(f"No numeric rate column found in {path}; skipping {cur}")
        rate_cache[cur] = None
        return None
    df_r = df_r.dropna(subset=[date_col, rate_col]).sort_values(date_col)
    if FILL_MISSING_RATES and not df_r.empty:
        full_idx = pd.date_range(df_r[date_col].min(), df_r[date_col].max(), freq='D')
        df_r = df_r.set_index(date_col).reindex(full_idx).rename_axis(date_col)
        df_r[rate_col] = df_r[rate_col].ffill()
        df_r = df_r.reset_index().rename(columns={'index': date_col})
    rate_cache[cur] = df_r
    rate_col_cache[cur] = (date_col, rate_col)
    return df_r

def get_spot(cur, d, lookback=RATE_LOOKBACK_DAYS):
    df_r = load_rates(cur)
    if df_r is None: return None
    date_col, rate_col = rate_col_cache[cur]
    target_date = pd.to_datetime(d).normalize()
    row = df_r[df_r[date_col] == target_date]
    if not row.empty:
        return float(row.iloc[0][rate_col])
    for i in range(1, lookback+1):
        cand = target_date - pd.Timedelta(days=i)
        row = df_r[df_r[date_col] == cand]
        if not row.empty:
            return float(row.iloc[0][rate_col])
    return None

def week_expiry(date):
    date = pd.to_datetime(date).normalize()
    for i in range(7):
        cand = date + pd.Timedelta(days=i)
        if cand.weekday() == 4:
            return cand
    return None

def extract_ccy(underlying):
    if not isinstance(underlying, str) or len(underlying) < 3:
        return None
    return underlying[:3].upper()

def select_atm(df_opts, mapping, cur, expiry, spot):
    if not all(k in mapping for k in ['underlying','expiry','strike']):
        return None
    sub = df_opts[(df_opts[mapping['expiry']] == expiry)].copy()
    if sub.empty:
        return None
    sub['__ccy'] = sub[mapping['underlying']].apply(extract_ccy)
    sub = sub[sub['__ccy'] == cur]
    if sub.empty:
        return None
    sub[mapping['strike']] = pd.to_numeric(sub[mapping['strike']], errors='coerce')
    sub = sub.dropna(subset=[mapping['strike']])
    if sub.empty:
        return None
    sub['atm_dist'] = (sub[mapping['strike']] - spot).abs()
    return sub.sort_values('atm_dist').iloc[0]

trade_rows = []

if 'mover_tweets' not in globals():
    print("mover_tweets not defined; run previous cell.")
else:
    tweet_date_col = 'date' if 'date' in mover_tweets.columns else None
    if tweet_date_col is None:
        print("No 'date' column in mover_tweets; cannot proceed.")
    else:
        mover_tweets[tweet_date_col] = pd.to_datetime(mover_tweets[tweet_date_col], errors='coerce').dt.tz_localize(None)
        for idx, row in mover_tweets.iterrows():
            base_date = row[tweet_date_col]
            if pd.isna(base_date):
                continue
            if isinstance(row.get('normalized_currencies'), list):
                currencies = [c for c in row['normalized_currencies'] if c in target_currencies]
            elif isinstance(row.get('currencies_found'), list):
                currencies = [c for c in row['currencies_found'] if c in target_currencies]
            else:
                currencies = []
            for cur in currencies:
                expiry = week_expiry(base_date)
                if expiry is None or (expiry - base_date).days > 7:
                    continue
                spot_entry = get_spot(cur, base_date)
                if spot_entry is None:
                    continue
                call_row = select_atm(fx_calls, calls_map, cur, expiry, spot_entry)
                put_row  = select_atm(fx_puts,  puts_map,  cur, expiry, spot_entry)
                if call_row is None or put_row is None:
                    continue
                strike_call = float(call_row[calls_map['strike']]) if 'strike' in calls_map else spot_entry
                strike_put  = float(put_row[puts_map['strike']]) if 'strike' in puts_map else spot_entry
                spot_expiry = get_spot(cur, expiry)
                if spot_expiry is None:
                    continue
                payoff = max(spot_expiry - strike_call, 0) + max(strike_put - spot_expiry, 0)
                trade_rows.append({
                    'tweet_id': row.get('id', row.get('tweet_id', idx)),
                    'currency': cur,
                    'tweet_date': base_date.normalize(),
                    'expiry': expiry,
                    'spot_entry': spot_entry,
                    'spot_expiry': spot_expiry,
                    'strike_call': strike_call,
                    'strike_put': strike_put,
                    'premium_call': FIXED_PREMIUM_PER_LEG,
                    'premium_put': FIXED_PREMIUM_PER_LEG,
                    'total_cost': 2*FIXED_PREMIUM_PER_LEG,
                    'payoff': payoff,
                    'return_pct': (payoff - 2*FIXED_PREMIUM_PER_LEG)/(2*FIXED_PREMIUM_PER_LEG) if 2*FIXED_PREMIUM_PER_LEG>0 else None,
                    'sentiment_compound': row.get('sentiment_compound'),
                    'tweet_text': (row.get('text','')[:120] + '...') if isinstance(row.get('text'), str) and len(row.get('text'))>120 else row.get('text')
                })

trades_df = pd.DataFrame(trade_rows)
print(f"Constructed {len(trades_df)} strangle trades (fixed premium {FIXED_PREMIUM_PER_LEG} per leg).")

if not trades_df.empty:
    print("Summary (by currency):")
    print(trades_df.groupby('currency')['return_pct'].agg(['count','mean','median']).round(4))
    print(f"\nOverall avg return: {trades_df['return_pct'].mean():.4%}")
    display(trades_df.head(20))
    trades_df.to_csv('long_strangle_backtest.csv', index=False)
    print("Saved long_strangle_backtest.csv")
else:
    print("No trades generated - still missing alignment or rate coverage causing spot/option selection skips.")
```
